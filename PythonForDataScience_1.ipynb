###Problem Solving

**Problem1:**
Given a list of numbers i.e [1,4,2,-5,9,7,3,2], make another list that contains all items in sorted order i.e [-5,1,2,2,3,4,7,9]

Approach: Selection Sort where first we will find th minimum value and swap it with the first value

```python
L =  [1,4,2,-5,9,7,3,2]
min_val = L[0]
idx = 0
count = 0
for i in L:
  if i < min_val:
    min_val = i
    idx = count
  count += 1
print(f"Minimum Value: {min_val} is found at index {idx}")
```

```python
temp = L[0]
L[0] = min_val
L[idx] = temp
print(L) #Now the minimum value is swapped with the first value
```

```python
L =  [1,4,2,-5,9,7,3,2]
for j in range(len(L)):
  min_val = L[j]
  idx = j
  count = j

  for i in range(j,len(L)):
    if L[i] < min_val:
      min_val = L[i]
      idx = count
    count += 1

  temp = L[j]
  L[j] = min_val
  L[idx] = temp

print(L) #Sorted List
```

Iterative Method for power of two

```python
def myPow(x: float, n: int) -> float:
    if n == 0:
        return 1
    if n < 0:
        x = 1 / x
        n = -n

    result = 1
    while n > 0:
        if n % 2 == 1:
            result *= x
        x *= x
        n //= 2
    return result
a = 3
b = 4
print(myPow(a,b))
```

Recursive Method for power of two

```python
def myPow(x: float, n: int) -> float:
    def power(x, n):
        if n == 0:
            return 1
        half = power(x, n // 2)
        return half * half if n % 2 == 0 else half * half * x

    if n < 0:
        x = 1 / x
        n = -n
    return power(x, n)
a = 3
b = 4
print(myPow(a,b))
```

**Arbitrary Number of Input Arguments in Python Functions**

In Python, you can create functions that accept any number of arguments using:

1. `*args` for **non-keyword(positional) arguments**

2. `*kwargs` for **keyword arguments**

###**1. *args ‚Äì Arbitrary Positional Arguments**

* Allows a function to accept any number of positional arguments.

* Inside the function, args is a tuple.

```python
#Syntax:
def function_name(*args):
    # args is a tuple of all positional arguments
    pass
```

```python
def mySum(*args):
  sum = 0
  for i in range(len(args)):
    sum += args[i]
  return sum
```

```python
mySum(4,3,2,1)
```

```python
mySum(2,9,5,5.9,78)
```

###**2. **kwargs ‚Äì Arbitrary Keyword Arguments**

* Allows a function to accept any number of keyword arguments.

* Inside the function, kwargs is a dictionary.

```python
#Syntax:
def function_name(**kwargs):
    # kwargs is a dictionary of all keyword arguments
    pass
```

```python
def AllVar(**kwargs):
  for x in kwargs:
    print("Variable Name is: ",x," And Value is: ",kwargs[x])
```

```python
AllVar(a = 3, b = "B", c = "CCC", y = 6.7)
```

```python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Teju", age=21, country="India")
```

```python
"""Let say you are a teacher and you have different student
records containing id of a student and the marks list in each subject
where different students have taken different number of subjects. All
these records are in hard copy. You want to enter all the data in computer
and want to compute the average marks of each student and display"""

def getDataFromUser():
    D = {}
    while True:
        studentId = input("Enter student ID: ")
        marksList = input("Enter the marks by comma separted values: ")
        moreStudents = input('Enter "no" to quit insertion: ')
        if studentId in D:
            print(studentId, "is already insterted")
        else:
            D[studentId] = marksList.split(",")
        if moreStudents.lower() == "no":
            return D
```

```python
studentData = getDataFromUser()
```

```python
studentData
```

```python
def getAvgMarks(D):
    avgMarks = {}
    for x in D:
        L = D[x]
        s = 0
        for marks in L:
            s += int(marks)
        avgMarks[x] = s/len(L)
    return avgMarks
```

```python
avgM = getAvgMarks(studentData)
```

```python
avgM
```

```python
for x in avgM:
    print("Student :",x,"got avg Marks as: ",avgM[x])
```
