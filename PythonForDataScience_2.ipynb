###`loc[]` → Label-based indexing

* Used to access rows and columns by labels/names.

* Syntax: `df.loc[row_label, column_label]`

```python
A.loc[1:3] #explicit indices
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Age': [25, 30, 35]}

df = pd.DataFrame(data, index=['a', 'b', 'c'])

print(df.loc['a'])        # Row with label 'a'
print("\n")
print(df.loc['a':'b'])    # Rows from 'a' to 'b' (inclusive)
print("\n")
print(df.loc[:, 'Name'])  # All rows, only 'Name' column
```

###` iloc[]` → Integer-location based indexing

* Used to access rows and columns by position / index

* Syntax: `df.iloc[row_index, column_index]`

```python
print(df.iloc[0])        # First row (index 0)
print("\n")
print(df.iloc[0:2])      # First two rows (0 and 1)
print("\n")
print(df.iloc[:, 0])     # All rows, first column
```

```python
A.iloc[1:3] #implicit indices
```

Comparison Table:

| Feature          | `.loc[]`            | `.iloc[]`               |
| ---------------- | ------------------- | ----------------------- |
| Indexing Type    | Label-based         | Position-based          |
| Row Access       | `df.loc['a']`       | `df.iloc[0]`            |
| Column Access    | `df.loc[:, 'Name']` | `df.iloc[:, 0]`         |
| Supports Slicing | Yes (inclusive)     | Yes (exclusive of stop) |
| Boolean Indexing | Yes                 | Yes                     |

**NOTE**

####Why we use `inplace` ?

In Pandas, the `inplace=True` parameter is used to modify the DataFrame directly, without creating a new copy.

Normally, many Pandas operations **return a new DataFrame**, leaving the original unchanged:

`df_new = df.drop('column_name', axis=1)` # Original df remains unchanged

But when you use:

`df.drop('column_name', axis=1, inplace=True)`

It modifies `df` itself, and returns `None`

####When to Use `inplace=True`

| Operation       | Description                |
| --------------- | -------------------------- |
| `drop()`        | Remove rows/columns        |
| `fillna()`      | Fill missing values        |
| `replace()`     | Replace values             |
| `sort_values()` | Sort rows by column values |
| `rename()`      | Rename rows/columns        |
| `reset_index()` | Reset the DataFrame index  |


Example:

Without `inplace`:

```python
df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = df1.drop('A', axis=1)   # df1 is unchanged
df2
```

```python
df1
```

With `inplace`:

```python
df1.drop('A', axis=1, inplace=True)  # df is modified directly
df1
```

####**When not to use `inplace`:**

Avoid `inplace` when:

* You want to retain the original DataFrame.

* You're working in a pipeline like:

`df_cleaned = df.dropna().sort_values('column')`
